```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Continuous plots 
df = pd.read_excel('z_values_sigmas.xlsx', sheet_name='Sheet1')
columns = df.columns
plt.figure(figsize=(8, 6))
for col_it in columns:  
        plt.plot(df[col_it], label=f'Instance {col_it}')

        plt.xlabel('Generations')
        plt.ylabel('Objective function')
        plt.legend(loc=1, title='Instances')
        plt.savefig(f'instances_a.pdf');
```
