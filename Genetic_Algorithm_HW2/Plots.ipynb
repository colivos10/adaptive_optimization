```python
seed_exp = [i for i in range(9)]
pbb_mutation_exp = [0.4]
population_size_exp = [100]
for pop_size_it in population_size_exp:
    for pbb_it in pbb_mutation_exp:
        plt.figure(figsize=(8, 6))
        for seed_it in seed_exp: 
            plt.plot(df['Generation'], 
                     df[f'{seed_it}_100_{pop_size_it}_{pbb_it}'], label=f'Seed {seed_it}')

        plt.xlabel('Generations')
        plt.ylabel('Objective function')
        plt.legend(loc=1, title='Instances')
        #plt.savefig(f'qap_initial_solution_effect.pdf');
        #plt.show()
```

```python
# Effect of the probability 
gen_size = 500
df = pd.read_excel(f'qap_ga_{gen_size}.xlsx')
seed_exp = [5]
pbb_mutation_exp = [0.1, 0.4, 0.6]
population_size_exp = [100]

for pop_size_it in population_size_exp:
    for seed_it in seed_exp:
        plt.figure(figsize=(8, 6))
        for pbb_it in pbb_mutation_exp:
            plt.plot(df['Generation'],df[f'{seed_it}_{gen_size}_{pop_size_it}_{pbb_it}'], label=f'$P_m = {pbb_it}$')
        
        plt.xlabel('Generations')
        plt.ylabel('Objective function')
        plt.legend(loc=1, title='Mutation Probability')
        plt.savefig(f'qap_pbb_effect_{seed_it}.pdf');
        #plt.show()
```

```python
exp_seeds = [0, 3, 5, 7, 10]
exp_temp = [100, 500, 1000]
exp_stops = [1000, 5000, 10000]
```

```python
for it_temp in exp_temp:
    for it_stops in exp_stops:
        plt.figure(figsize=(8, 6))
        for it_seed in exp_seeds:
            df = pd.read_excel(f'results_qap/exp_{it_seed}_{it_temp}_{it_stops}.xlsx')
            plt.plot(df['temp'], df['cost'], label=f'Seed {it_seed}')
        plt.xlim(max(df['temp']), 0)
        plt.xlabel('Temperature')
        plt.ylabel('Objective function')
        plt.legend(loc=1, title='Instances')
        plt.savefig(f'plots_comb/{it_temp}_{it_stops}.pdf');
        #plt.show()
```

```python
for it_seed in exp_seeds:
    for it_stops in exp_stops:
        plt.figure(figsize=(8, 6))
        for it_temp in exp_temp:
            df = pd.read_excel(f'results_qap/exp_{it_seed}_{it_temp}_{it_stops}.xlsx')
            plt.plot(df['cost'], label=f'Temperature  {it_temp}')
        plt.xlabel('Iterations')
        plt.ylabel('Objective function')
        plt.legend(loc=1, title='Instances')
        plt.savefig(f'plots_comb/cost_{it_seed}_{it_stops}.pdf');
        #plt.show()
```
