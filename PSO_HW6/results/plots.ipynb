```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# v1 
df = pd.read_excel('best_objectives_v1.xlsx')
columns = df.columns
plt.figure(figsize=(8, 6))
for col_it in columns:  
        plt.plot(df[col_it], label=f'Instance {col_it}')
        plt.xlabel('Iterations')
        plt.ylabel('Objective function')
        plt.legend(loc=1, title='Instances')
plt.savefig(f'question_egg1.pdf');
plt.figure(figsize=(8, 6))
df.boxplot(column=[f'Seed {i}' for i in range(10)])
plt.savefig(f'question_egg1_boxplot.pdf');
```
