```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Continuous plots 
df = pd.read_excel('results_cont/exp_0_50_1000.xlsx')
df.head()
```

```python
exp_seeds = [0, 3, 5, 7, 10]
exp_temp = [50, 100, 500]
exp_stops = [1000, 5000, 10000]
```

```python
for it_temp in exp_temp:
    for it_stops in exp_stops:
        plt.figure(figsize=(8, 6))
        for it_seed in exp_seeds:
            df = pd.read_excel(f'results_cont/exp_{it_seed}_{it_temp}_{it_stops}.xlsx')
            plt.plot(df['temp'], df['cost'], label=f'Seed {it_seed}')
        plt.xlim(max(df['temp']), 0)
        plt.xlabel('Temperature')
        plt.ylabel('Objective function')
        plt.legend(loc=1, title='Instances')
        plt.savefig(f'plots_cont/{it_temp}_{it_stops}.pdf');
        #plt.show()
```

```python
for it_seed in exp_seeds:
    for it_stops in exp_stops:
        plt.figure(figsize=(8, 6))
        for it_temp in exp_temp:
            df = pd.read_excel(f'results_cont/exp_{it_seed}_{it_temp}_{it_stops}.xlsx')
            plt.plot(df['cost'], label=f'Temperature  {it_temp}')
        plt.xlabel('Iterations')
        plt.ylabel('Objective function')
        plt.legend(loc=1, title='Instances')
        plt.savefig(f'plots_cont/cost_{it_seed}_{it_stops}.pdf');
        #plt.show()
```
